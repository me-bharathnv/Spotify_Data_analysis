# filtering the data

```python
import pandas as pd
data = pd.read_csv("SpotifyFeatures.csv")
data
```

```python
data.loc[(data["genre"]=="Movie") & (data["popularity"] > 50)]  # filtering the data using loc funciton
```

```python
data_soul = data.loc[(data["genre"]=="Soul") & (data["popularity"] > 50)] # filtering the data using loc and or funciton
data_soul = data_soul.reset_index()
data_soul.to_csv("Spotify_soul.csv") # saving the file in csv
```

# aggregate Stats

```python
import pandas as pd
data = pd.read_csv("SpotifyFeatures.csv")
print(data.columns)
```

```python
data.groupby(["genre"]).mean().sort_values("popularity")  # group by the genre and sor values by popularity
```

```python
data.groupby(["genre"]).sum().sort_values("popularity") # sum of the popularity
```

```python
data.groupby(["genre"]).count().sort_values("popularity") # count the number of genre by popularity
```
